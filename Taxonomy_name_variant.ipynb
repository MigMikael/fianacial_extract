```python
import re
import os
import json
import pandas as pd
```

```python
quarter = 'Q2'
form_no = 'F6'
year_list = ['2022', '2023']
# form_sym_list = ['BAY', 'BAM', 'SCB', 'CIMBT', 'LHFG']
# form_sym_list = ['JMT', 'AIRA', 'SAWAD', 'AEONTS', 'ACAP']
# form_sym_list = ['MFC', 'KGI', 'MST', 'BYD', 'XPG']
# form_sym_list = ['TSI', 'MTI', 'TIPH', 'NKI', 'BUI']
# form_sym_list = ['MJLF', 'FTREIT', 'BKKCP', 'BTSGIF', '3BBIF']
form_sym_list = ['SCC', 'BTS', 'TKN', 'OCC', 'BANPU', 'WINDOW']

dtype = {
    'account_name': 'string',
    'taxonomy_account_name': 'string',
    'account_code': 'string',
    'values': 'string'
}
```

```python
listed_df_list = []
for item in form_sym_list:
    for year in year_list:
        mapping_file_path = f"./Listed_Mapping_Prod/{year}/{item}_{quarter}_{year}.csv"
        print(mapping_file_path)

        if (os.path.isfile(mapping_file_path)):
            new_listed_mapping_df = pd.read_csv(mapping_file_path, encoding='utf-8', dtype=dtype)
            listed_df_list.append(new_listed_mapping_df)
```

```python
all_listed_df = pd.concat(listed_df_list, ignore_index=True)
```

```python
len(all_listed_df.index)
```

```python
taxonomy_file_path = f"./SET_Taxonomy/SET_TAXONOMY_V3/Taxonomy_{form_no}_V3.csv"
output_taxonomy_file_path = f"./SET_Taxonomy/SET_TAXONOMY_V4/Taxonomy_{form_no}_V4_rev1.csv"

dtype = {
    'form_no': 'string',
    'set_code': 'string',
    'dm_code': 'string',
    'running_number': 'string',
    'reference_no': 'string',
    'indent': 'string',
    'item_in_taxonomy': 'string',
    'item_in_taxonomy_en': 'string',
    'parent_code': 'string',
    'parent_code_operation': 'string',
    'formating': 'string',
    'data_type': 'string',
    'symbol': 'string',
    'financial_statement_type': 'string',
    'fs_category': 'string',
    'description_th': 'string',
    'description_en': 'string',
    'opposite_code_short': 'string',
    'mandatory_mapping': 'string',
    'zone_code': 'string',
    'warning_mapping_msg_th': 'string',
    'warning_mapping_msg_en': 'string',
    'meaning_of_taxonomy': 'string',
}
taxonomy_df = pd.read_csv(taxonomy_file_path, encoding='utf-8', dtype=dtype, na_filter=False)
taxonomy_df.head()
```

```python
count_add_meaning = 0
for index, row in taxonomy_df.iterrows():
    if (row['data_type'] == 'T'):
        continue

    # taxonomy_df.at[index, 'meaning_of_taxonomy'] = ""
    
    set_code = row['set_code']
    set_name = row['item_in_taxonomy']
    
    match_row_df = all_listed_df[all_listed_df['account_code'] == set_code]

    fs_name_list = []
    for i, r in match_row_df.iterrows():
        acc_name = r['account_name']
        if (acc_name != set_name):
            fs_name_list.append(acc_name)
            
    fs_name_set = set(fs_name_list)
    fs_name_list = list(fs_name_set)

    fs_name_str = " หรือ ".join(fs_name_list)
    fs_name_str = re.sub(r'\{{\w+\}}', '', fs_name_str)
    fs_name_str = "บางครั้งใช้ชื่อว่า " + fs_name_str

    if (len(fs_name_list) > 0):
        count_add_meaning += 1
        meaning = row['meaning_of_taxonomy']
        taxonomy_df.at[index, 'meaning_of_taxonomy'] = fs_name_str + str(meaning)
        
        print(set_code, set_name)
        print(len(match_row_df.index))
        print(fs_name_list)
        print(len(fs_name_list))
        print(fs_name_str)
        print("\n\n")
```
